## Imports

```python
import pandas as pd
from IPython.display import IFrame
import matplotlib.pyplot as plt
```

## Dataset

##### *Features*

```python
dt = pd.read_csv('../datamart/data_project_nel.csv')
dt
```

```python
dt.columns
```

```python
dt.describe()
```

Check for missing values

```python
dt.isna().sum()
```

Almost half of the data in the dry_days variable is null, so we will drop this variable

```python
dt.drop(columns=['dry_days'], inplace=True)
```

Basic exploratory plots

```python
dt.hist(figsize=(15, 15), bins=50)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
dt.plot(kind='box', figsize=(15, 8))
plt.xticks(ticks=range(1, len(dt.columns) + 1), labels=dt.columns, rotation=45)
plt.tight_layout()
plt.show()
```

We decided to scale the data using RobustScaler, due to outliers present in the dataset

```python
from sklearn.preprocessing import RobustScaler

transformer = RobustScaler().fit(dt)
robust_scaled_data = transformer.transform(dt)
robust_df = pd.DataFrame(robust_scaled_data, columns = dt.columns)
robust_df
```

```python
robust_df.plot(kind='box', figsize=(15, 8))
plt.xticks(ticks=range(1, len(robust_df.columns) + 1), labels=robust_df.columns, rotation=45)
plt.tight_layout()
plt.show()
```

Correlation matrix

```python
import seaborn as sns

correlation_matrix = robust_df.corr()
plt.figure(figsize=(20, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```

We chose to drop delivery_age_years as it has 0.99 correlation with lactation

```python
robust_df.drop(columns=['delivery_age_years'], inplace=True)
```
